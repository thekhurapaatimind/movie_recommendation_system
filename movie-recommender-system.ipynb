```python
import pandas as pd
import numpy as np
```

```python
movies =  pd.read_csv('Data/tmdb_5000_movies.csv')
credits =  pd.read_csv('Data/tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.info()
```

```python
movies = movies[['genres', 'id', 'keywords', 'overview', 'title', 'cast', 'crew']]
```

```python
movies.head(2)
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
movies.head(1)
```

```python
movies.iloc[0]['crew']
```

```python
def fetch_director(text):
    for i in ast.literal_eval(text):
        if(i['job'] == 'Director'):
            return i['name']
    
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def condense(L):
    L1 = []
    for i in L:
        L1.append(i.replace(' ',''))
    return L1
```

```python
movies.dropna(inplace=True)
```

```python
movies.info()
```

```python
movies['cast'] = movies['cast'].apply(condense)
movies['genres'] = movies['genres'].apply(condense)
movies['keywords'] = movies['keywords'].apply(condense)
```

```python
movies.sample(5)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast'])
```

```python
new.head()
```

```python
#new.head()
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new['tags'] = new['tags'] + new['crew']
new.sample(5)
```

```python
new = new.drop(columns=['crew'])
new.sample(2)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    L = []
    for i in text.split():
        L.append(ps.stem(i))
    return " ".join(L)
```
